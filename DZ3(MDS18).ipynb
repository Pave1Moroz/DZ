```python
import numpy as np
import matplotlib.pyplot as plt
```

1.

```python
A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
B = np.array([[0.5,0,0],[0,0.5,0],[0,0,0.5]])
C = B.dot(A.transpose())

D = C.transpose()
x = D[:,0]
y = D[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
