```python
def func(x):
    return np.sin(2*x+1)**5

sin(2*x+1)**5
```

```python
def first_deriv_func(x):
    return 10*np.cos(2*x+1)*np.sin(2*x+1)**4

10*cos(2*x+1)*sin(2*x+1)**4 #первая производная
```

```python
def second_deriv_func(x):
    return 80*np.sin(2*x+1)**3*np.cos(2*x+1)**2-20*np.sin(2*x+1)**5

80*sin(2*x+1)**3*cos(2*x+1)**2-20*sin(2*x+1)**5  #вторая
```

```python
#Исходная функция
xlist = []
ylist = []
x = -5
for a in range(100):
    xlist.append(x)
    ylist.append(func(x))
    x+=0.1
    
plt.plot(xlist, ylist, label='func')
plt.xlim([-6, 6])
```

```python
#Первая производная
xlist = []
ylist = []
x = -5
for a in range(100):
    xlist.append(x)
    ylist.append(first_deriv_func(x))
    x+=0.1
    
plt.plot(xlist, ylist, label='func')
plt.xlim([-6, 6])
```

```python
#Первая производная
xlist = []
ylist = []
x = -5
for a in range(100):
    xlist.append(x)
    ylist.append(second_deriv_func(x))
    x+=0.1
    
plt.plot(xlist, ylist, label='func')
plt.xlim([-6, 6])
```
