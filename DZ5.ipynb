```python
#ВВОДНЫЕ ДАННЫЕ:

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Владимир Иванов'}]

directories = {
 '1': ['2207 876234', '11-2', '10006'],
 '2': ['10006'],
 '3': []}
```

```python
#ВСПОМОГАТЕЛЬНЫЕ ФУНКЦИИ :

def inputing(input_string):
    """
    Ввод по запросу
    """
    enter = str(input(input_string))
    return enter

def printing(result_list, sting_none, string_one, string_many):
    """
    Вывод списка результатов на экран
    """
    if len(result_list) < 1: print(sting_none)
    elif len(result_list) < 2: print(string_one, *result_list)
    else:
        print(string_many)
        for result in result_list: print(result)
            
def dr():
    """
    Вывод списка полок
    """
    print('Список полок:')
    for directory in directories:
        for number in directory:
            print('№:', directory , 'Номера документов на полке:', ', '.join(directories[number]))
            
def com():
    print(' Команда: «com» - Список команд', '\n',
          'Команда: «p» Действие: Возвращает имя/имена владельцев документов с заданным на вход номером', '\n',
          'Команда: «s» Действие: Возвращает номер полки, на которой находится документ с заданным на вход номером', '\n',
          'Команда: «l» Действие: Вывод полной информации по всем документам', '\n',
          'Команда: «dr» Действие: Вывод списка полок', '\n',
          'Команда: «ads» Действие: Добавление новой полки в список полок', '\n',
          'Команда: «ds» Действие: Удаление полки из списка полок', '\n',
          'Команда: «ad» Действие: Добавление документа в данные', '\n',
          'Команда: «d» Действие: Удаление всех документов из данных, имеющих заданный номер', '\n',
          'Команда: «m» Действие: Перемещение документа с одной полки на другую')
    
#ЗАДАНИЕ 1:

def p(document_number):
    """
    Возвращает имя/имена владельцев документов с заданным на вход номером
    """
    matching_document_owners = []
    for document in documents:
        if document['number'] == document_number:
            matching_document_owners.append(document['name'])
    return matching_document_owners

#ЗАДАНИЕ 2:

def s(document_number):
    """
    Возвращает номер полки, на которой находится документ с заданным на вход номером
    """
    matching_directories = []
    for directory, numbers in directories.items():
        for number in numbers:
            if number == document_number:
                matching_directories.append(directory)
    return matching_directories

#ЗАДАНИЕ 3:

def l():
    """
    Вывод полной информации по всем документам
    """
    for document in documents:
        print('№: ', document['number'], 'тип: ', document['type'], 
              'Владелец: ', document['name'], 'Полки хранения: ', ', '.join(s(document['number'])))

#ЗАДАНИЕ 4:

def ads(directory_number):
    """
    Добавление новой полки в список полок
    """
    directory_exist = False
    for directory in directories:
        if directory == directory_number: directory_exist = True
    if directory_exist == True:
        print('Такая полка уже существует. Текущий перечень полок: ', ', '.join(directories))
    else:
        directories[directory_number] = []
        print('Полка добавлена. Текущий перечень полок: ', ', '.join(directories))
        
#ЗАДАНИЕ 5:

def ds(directory_number):
    """
    Удаление полки из списка полок
    """
    if directory_number in directories:
        if len(directories[directory_number]) > 0:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', ', '.join(directories))
        else:
            del(directories[directory_number])
            print('Полка удалена. Текущий перечень полок: ', ', '.join(directories))
    else: print('Такая полка не существует. Текущий перечень полок: ', ', '.join(directories))
        
#ЗАДАНИЕ 6:

def ad():
    """
    Добавление документа в данные
    """
    new_document = {}
    new_document['number'] = inputing('Введите номер документа: ')
    new_document['type'] = inputing('Введите тип документа: ')
    new_document['name'] = inputing('Введите владельца документа: ')
    new_document_rack = inputing('Введите полку для хранения: ')
    if new_document_rack in directories:
        if new_document['number'] in directories[new_document_rack]:
            print(' На этой полке уже есть документ с таким номером.', '\n', 
                  'Добавьте документ на другую полку или создайте новую командой ads.', '\n', 'Текущий список документов: ')
            l()
        else:
            if new_document_rack in directories:
                documents.append(new_document)
                directories[new_document_rack].append(new_document['number'])
                print('Документ добавлен. Текущий список документов: ')
                l()
    else: 
        print('Такой полки не существует. Добавьте полку командой as. Текущий список документов: ')
        l()
            
#ЗАДАНИЕ 7:

def d(document_number):
    """
    Удаление всех документов из данных, имеющих заданный номер
    """
    del_count = 0
    delete_document = []
    for document in documents:
        if document['number'] == document_number:
            delete_document.append(documents.index(document))
            
    offset = 0
    for ind in delete_document:
        del(documents[ind-offset])
        offset += 1
        del_count += 1
            
    for directory in directories.values():
        for number in directory:
            if number == document_number:
                del(directory[directory.index(number)])
                
    if del_count > 0:
        print(' Удалено документов: ', del_count, '\n', 'Текущий список документов: ')
        l()
    else: 
        print('Документ не найден в базе', '\n', ' Текущий список документов: ')
        l()
        
#ЗАДАНИЕ 8:

def d(document_number):
    """
    Удаление всех документов из данных, имеющих заданный номер
    """
    del_count = 0
    delete_document = []
    for document in documents:
        if document['number'] == document_number:
            delete_document.append(documents.index(document))
            
    offset = 0
    for ind in delete_document:
        del(documents[ind-offset])
        offset += 1
        del_count += 1
            
    for directory in directories.values():
        for number in directory:
            if number == document_number:
                del(directory[directory.index(number)])
                
    if del_count > 0:
        print(' Удалено документов: ', del_count, '\n', 'Текущий список документов: ')
        l()
    else: 
        print('Документ не найден в базе', '\n', ' Текущий список документов: ')
        l()
        
#ЗАДАНИЕ 9:

def m(document_number):
    """
    Перемещение документа с одной полки на другую
    """
    directory_list = s(document_number) #список полок, на которых находится документ с искомым номером
    if len(directory_list) == 0:
        print(' Документ не найден в базе', '\n', ' Текущий список документов: ')
        l()
    else:
        new_rack = inputing('Новый номер полки? : ')
        if new_rack in directories:
            if document_number in directories[new_rack]:
                print(' На новой полке уже есть документ с таким номером.', '\n', ' Текущий список документов: ')
                l()
            else:
                #Условие на случай, если есть несколько полок с таким номером документа:
                if len(directory_list) > 1:
                    #непонятно какой именно документ надо переместить:
                    print(' Найдено насколько полок с таким номером документа: ', *directory_list)
                    old_rack = inputing('С какой именно полки нужно переместить документ? : ')
                    del(directories[old_rack][directories[old_rack].index(document_number)])
                    directories[new_rack].append(document_number)
                    print(' Документ перемещён.', '\n', ' Текущий список документов: ')
                    l()
                else:   
                #Цикл для удаления номера документа из "old_rack":
                    for directory in directories.values():
                        for number in directory:
                            if number == document_number:
                                del(directory[directory.index(number)])
                    directories[new_rack].append(document_number)
                    print(' Документ перемещён.', '\n', ' Текущий список документов: ')
                    l()
        else:
            print(' Такой полки не существует.', '\n', ' Текущий список документов: ')
            l()
            
#Главная функция:

def main(command):
    if command == 'p':
        printing(p(inputing('Номер документа? : ')), 'Документ не найден в базе', 'Владелец документа: ', 
                 'Найдено несколько человек с таким номером документа:')
    elif command == 's':
        printing(s(inputing('Номер документа? : ')), 'Документ не найден в базе', 'Номер полки: ', 
                 'Найдено несколько полок с документами по этому номеру: ')
    elif command == 'l':
        l()
    elif command == 'dr':
        dr()
    elif command == 'ads':
        ads(inputing('Номер полки? : '))
    elif command == 'ds':
        ds(inputing('Номер полки? : '))
    elif command == 'ad':
        ad()
    elif command == 'd':
        d(inputing('Номер документа? : '))
    elif command == 'm':
        m(inputing('Номер документа? : '))
    elif command == 'com':
        com()
    else:
        print('Такой комманды не существует. Получить список команд можно, введя «com».')
```

```python
main(inputing('Имя команды? : '))
```
