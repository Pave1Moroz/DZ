```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from scipy.optimize import minimize

def rastr(X):
    a=10.0
    return a*len(X) + sum([(x**2 - a * np.cos(2 * np.pi * x)) for x in X])
```

.

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(45, 30)

X = np.arange(-5.12, 5.12, 0.01)
y = np.arange(-5.12, 5.12, 0.01)
X, y = np.meshgrid(X, y)
z = rastr(np.array([X, y]))

surf = ax.plot_surface(X, y, z, cmap=cm.coolwarm)
plt.show()
```

.

```python
#  2

itList = []
resList = []

for i in range(100):
    x0=np.random.uniform(low=0, high=2, size=(2,))
    res = minimize(rastr, x0, method='nelder-mead', options={'xtol': 1e-9, 'disp': False})
    itList.append(res.nit)
    resList.append(rastr(res.x))
    
print(' : ', np.mean(resList), 
      ' : ', np.min(resList), '  : ', np.mean(itList))
```
